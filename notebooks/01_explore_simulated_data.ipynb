```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", context="notebook")
```

```python
df = pd.read_csv("../data_examples/final_simulated_panel.csv")
df.head()
```

```python
df.info()
df.describe().T
```

```python
sns.histplot(df["sales"], bins=50)
plt.title("Distribution of Sales")
plt.show()
```

```python
# show sales for one randomly selected store
store = df['store_id'].sample(1).iloc[0]

temp = df[df.store_id == store]

sns.lineplot(data=temp, x="week", y="sales")
plt.title(f"Sales Over Time for Store {store}")
plt.show()
```
